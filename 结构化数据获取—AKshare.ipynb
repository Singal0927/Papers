### 一、市场数据

```python
def obtainData(symbol,start='20210226',end='20220212'):
    data=ak.stock_zh_a_hist(symbol=symbol,period='daily',start_date=start,end_date=end)
    data=data.set_index('日期')
    ror=pd.DataFrame(np.log(data['收盘']/data['收盘'].shift()))
    ror.columns=[symbol]
    ror=ror.drop(['2021-02-26'])#drop用于删除行（序号）或者列（列名）
    return ror
```

### 二、动态风险数据
获取个别银行的损失变量：

```python
def obtainLoss(symbol,start='20210226',end='20220212'):
    data=ak.stock_zh_a_hist(symbol=symbol,period='daily',start_date=start,end_date=end)
    data=data.set_index('日期')
    loss=pd.DataFrame(data['涨跌幅'])
    loss=-loss
    loss.columns=[symbol]
    loss.drop(['2021-02-26'])
    return loss
```

```python
symbols=['601398','601939','601288','601988','601658','601328','600036','600000','601166','601998','600016','601818','000001','600015','601169','601229','600919','601916','002142','601009']

df3=obtainLoss('601398')
for symbol in symbols[1:]:
    df4=obtainLoss(symbol)
    df3=pd.concat([df3,df4],axis=1)
df3
```

获取银行系统的损失变量：将各家银行损失率的加权平均值作为银行系统的损失变量：

```python
values=[]
for symbol in symbols:
    info=ak.stock_individual_info_em(symbol=symbol)
    value=info['value'][info['item']=='总市值'].item()
    values.append(value)

values=np.array(values)
values/=values.sum()
```

```python
df3['sys']=list(round((values*df3).mean(axis=1),2))
```

```python
df3.to_csv('./数据/动态风险数据.csv',encoding='utf_8_sig')
```

获取状态变量：所有变量的index均必须DatetimeIndex类型，如此才能使得concat函数根据index进行拼接。

```python
x1_2 = ak.bond_china_yield(start_date="2021-02-24", end_date="2022-02-11")
x1_2=x1_2.set_index('日期')
x1_2.index=pd.DatetimeIndex(x1_2.index)

x1=x1_2['3月'][x1_2['曲线名称']=='中债国债收益率曲线']

x2=x1_2['10年'][x1_2['曲线名称']=='中债国债收益率曲线']-x1_2['3月'][x1_2['曲线名称']=='中债国债收益率曲线']
```

```python
x1_2.index
```

```python
x1=x1-x1.shift()#国债到期收益率的变化，采用3月期国债到期收益率的一阶差分来表示
x1
```

```python
x2#长短期利差，使用10年期国债到期收益率和3月期国债到期收益率之差作为代理变量
```

```python
x3 = ak.macro_china_shibor_all()
x3.index=pd.DatetimeIndex(x3.index)
x3=x3['3M_定价']['2021-02-24':'2022-02-10']
x3=x3-x1
x3#流动性利差，采用3月期上海银行间同业拆放利率(SHIBOR)和3月期国债到期收益率之差作为代理变量
```

```python
x4=x1_2['10年'][x1_2['曲线名称']=='中债中短期票据收益率曲线(AAA)']-x1_2['10年'][x1_2['曲线名称']=='中债国债收益率曲线']
x4.index=pd.DatetimeIndex(x4.index)
x4#信用利差的变化，信用利差用10年期AAA级企业债到期收益率和10年期国债到期收益率的差来表示
```

```python
x5_6=ak.stock_zh_index_daily(symbol='sh000001').set_index('date')
x5_6.index=pd.DatetimeIndex(x5_6.index)#索引转换为日期类型

x5=x5_6['2021-02-24':'2022-02-10']
x5['sh000001']=np.log(x5['close']/x5['close'].shift())
x5=x5.dropna()
x5=x5['sh000001']
x5#市场收益率，用上证综指每日涨跌幅的对数一阶差分来表示
```

```python
x6=x5_6['2021-01-19':'2022-02-10']
x6=(x6['close']-x6['close'].shift()).rolling(22).std().dropna()
x6#波动率，将上证综指日度涨跌幅的22天滚动标准差作为代理变量
```

```python
M=pd.concat([x1,x2,x3,x4,x5,x6],axis=1)
M.columns=['m1','m2','m3','m4','m5','m6']
M=M.dropna()
M
```

```python
M.to_csv('./数据/CoVaR的状态变量.csv')
```
